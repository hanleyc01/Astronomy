# Plotting out distances

For the sake of our own knowledge, and because Prof. said he hadn't seen it before, we'll be plotting out the 
distances of each of the Messier objects.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

messiers = pd.read_csv('Messier Catalog List - Sheet1.csv')
```

```python
plt.figure(figsize=(10, 20))
plt.plot(range(len(messiers)), messiers['DIST (ly)'])
plt.yticks(rotation=45)
plt.xlabel('Messier Object Num.')
plt.ylabel('Distance (Ly)')
plt.title('Messier Object Distance by Messier Number')
```

```python
plt.figure(figsize=(10, 20))
plt.plot(range(len(messiers)), messiers['MAG'])
plt.xlabel('Messier Object Num.')
plt.ylabel('Magnitude')
plt.title('Messier Object Magnitude by Messier Number')
```
